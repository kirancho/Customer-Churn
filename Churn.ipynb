#Feature Scaling

```python
#Save remaining continuous variables into cnames
cnames = [ 'account length', 'number vmail messages', 'total day minutes', 'total day calls',
          'total eve minutes', 'total eve calls', 'total night minutes', 'total night calls',
          'total intl minutes', 'total intl calls', 'number customer service calls' ]

#Loop to normalise the datasets
for i in cnames:
    TrainData[i] = (TrainData[i] - TrainData[i].min())/(TrainData[i].max() - TrainData[i].min())
    TestData[i] = (TestData[i] - TestData[i].min())/(TestData[i].max() - TestData[i].min())
```

#Modelling

```python
#Replace target cateogy by 'False' and 'True'
TrainData['Churn'] = TrainData['Churn'].replace(0, 'False')
TrainData['Churn'] = TrainData['Churn'].replace(1, 'True')
TestData['Churn'] = TestData['Churn'].replace(0, 'False')
TestData['Churn'] = TestData['Churn'].replace(1, 'True')
```

```python
# Define a funtion that returns Accuracy and FNR
def EvaluationMatrix(Conf_Mat):
    TN = Conf_Mat.iloc[0,0]
    FN = Conf_Mat.iloc[1,0]
    TP = Conf_Mat.iloc[1,1]
    FP = Conf_Mat.iloc[0,1]

    #Accuracy of model
    Accuracy = ((TP+TN)*100)/(TP+TN+FP+FN)

    #False Negative rate 
    FNR = (FN*100)/(FN+TP)
    return(Accuracy,FNR)
```

```python
#Decision Tree Classifier
DT_Cls = tree.DecisionTreeClassifier(class_weight = 'balanced', criterion = 'entropy').fit(TrainData.values[:,0:14], TrainData.values[:,14])
```

```python
#Predict test cases
DT_Cls_Predict = DT_Cls.predict(TestData.iloc[:,0:14])
```

```python
#Create a dot file to visualise tree
#www.webgraphviz.com
dot_file = open("tree_plt.dot", "w")
df = tree.export_graphviz(DT_Cls, out_file = dot_file, feature_names = TrainData.columns[:14])
```

```python
#Build Confusion Matrix
Conf_Mat_DT = pd.crosstab(DT_Cls_Predict, TestData.values[:,14])
```

```python
#Confusion matrix
Conf_Mat_DT
```

```python
#Find Accuracy & FNR
Accuracy_DT, FNR_DT = EvaluationMatrix(Conf_Mat_DT)
print("Accuracy of DT model is:", Accuracy_DT )
print("FNR of DT model is:", FNR_DT)
```
